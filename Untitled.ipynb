```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('Datasets/B2-2122.csv',delimiter=',')
```

```python
raw_data.head()
```

```python
# convert date and time into a pandas datetime object
raw_data["Date"] = pd.to_datetime(raw_data['Date'] + ' ' + raw_data['Time'],format='%d/%m/%Y %H:%M')
```

```python
#raw_data.drop(columns = ["Div"])
raw_data.head()
```

```python
# drop all betting stats and Div
soccer_stats = raw_data.loc[:,'Date':'AR']
```

```python
# no need for a seperate time column any more
soccer_stats = soccer_stats.drop(columns=["Time"])
```

```python
# add one hour to fix the time
soccer_stats["Date"] = soccer_stats["Date"] + pd.Timedelta(hours=1)
```

```python
# Get all teams in the 2. Bundesliga in sorted order
teams = np.sort(soccer_stats["HomeTeam"].unique())
teams
```

```python
import datetime
# soccer_stats[soccer_stats.Date.dt.time == datetime.time(20,30)]
satur = soccer_stats.query('Date.dt.time == datetime.time(20,30) & Date.dt.dayofweek == 5')
```

```python
home_satur = satur.groupby("HomeTeam").count()["Date"]
home_satur
```

```python
away_satur = satur.groupby("AwayTeam").count()["Date"]
```

```python
ser_teams = pd.Series(data = None, index = teams)
ser_teams
```

```python
satur_stats= pd.concat([home_satur,away_satur,ser_teams],axis=1,ignore_index=True)
satur_stats = satur_stats.drop(columns=[2])
satur_stats = satur_stats.fillna(0)
satur_stats = satur_stats.astype('int32')
satur_stats.rename(columns={0:"Heim",1:"Auswärts"},inplace=True)
# satur_stats.rename(columns= { satur_stats[0]:"Heim"})
satur_stats
# satur_stats
```

```python
satur_stats['Gesamt'] = satur_stats[list(satur_stats.columns)].sum(axis=1)
```

```python
satur_stats = satur_stats.sort_values(by=['Gesamt'],ascending=False)
```
